```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

# Load ideal data
ideal_data = pd.read_csv("/kaggle/input/store-transaction-data/Hackathon_Ideal_Data.csv")

# Handle missing values
imputer = SimpleImputer(strategy="mean")
ideal_data_filled = imputer.fit_transform(ideal_data.select_dtypes(include=['int', 'float']))
ideal_data_filled = pd.DataFrame(ideal_data_filled, columns=ideal_data.select_dtypes(include=['int', 'float']).columns)

# Print imputed data
print("Ideal Data after imputing missing values:")
print(ideal_data_filled.head())

# Data cleaning, normalization, and standardization
#scaler = StandardScaler()
scaler = MinMaxScaler()
ideal_data_scaled = scaler.fit_transform(ideal_data_filled.select_dtypes(include=['int', 'float']))
ideal_data_scaled = pd.DataFrame(ideal_data_scaled, columns=ideal_data_filled.select_dtypes(include=['int', 'float']).columns)

# Print scaled data
print("Ideal Data after normalization:")
print(ideal_data_scaled.head())

# Convert categorical variables into numerical representations (One-hot encoding)
encoder = OneHotEncoder()
ideal_data_encoded = pd.get_dummies(ideal_data.select_dtypes(include=['object']))

# Print encoded data
print("Ideal Data after one hot encoding:")
print(ideal_data_encoded.head())

# Concatenate numerical and encoded categorical features
ideal_data_preprocessed = pd.concat([ideal_data_scaled, ideal_data_encoded], axis=1)


# Print preprocessed data
print("Ideal Data after concatenation:")
print(ideal_data_preprocessed.head())
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

# Load working data
working_data = pd.read_csv("/kaggle/input/store-transaction-data/Hackathon_Working_Data.csv")

# Handle missing values
imputer = SimpleImputer(strategy="mean")
working_data_filled = imputer.fit_transform(working_data.select_dtypes(include=['int', 'float']))
working_data_filled = pd.DataFrame(working_data_filled, columns=working_data.select_dtypes(include=['int', 'float']).columns)

# Print imputed data
print("Working Data after imputing missing values:")
print(working_data_filled.head())

# Data cleaning, normalization, and standardization
#scaler = StandardScaler()
scaler = MinMaxScaler()
working_data_scaled = scaler.fit_transform(working_data_filled.select_dtypes(include=['int', 'float']))
working_data_scaled = pd.DataFrame(working_data_scaled, columns=working_data_filled.select_dtypes(include=['int', 'float']).columns)

# Print scaled data
print("Working Data after normalization:")
print(working_data_scaled.head())

# Convert categorical variables into numerical representations (One-hot encoding)
encoder = OneHotEncoder()
working_data_encoded = pd.get_dummies(working_data.select_dtypes(include=['object']))

# Print encoded data
print("Working Data after one hot encoding:")
print(working_data_encoded.head())

# Concatenate numerical and encoded categorical features
working_data_preprocessed = pd.concat([working_data_scaled, working_data_encoded], axis=1)


# Print preprocessed data
print("Working Data after concatenation:")
print(working_data_preprocessed.head())
```
