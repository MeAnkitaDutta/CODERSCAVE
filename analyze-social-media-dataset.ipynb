```python


def sentiment_calssification(x): 
    i = x
    if i < 2:
        return "Negetive"
    elif i > 2 & i < 4:
        return "Nutral"
    else:
        return "Positive"
        
        
df['Sentiment'] = df['Total_Sentiment'].apply(lambda x : sentiment_calssification(x) )
```

```python
temp = df.groupby("Time")["Sentiment"].value_counts().reset_index()

fig, ax = plt.subplots(1,1,figsize=(17,7),dpi=72)
sns.barplot(data=temp,x="Sentiment",y="count",hue="Time")
plt.title("Overall count of Sentiments", fontsize=15,color='grey',fontweight='bold')
plt.show()
```

```python
temp = df[["Positive"]]
# Get month for each dates in the index 
index_month = temp.index.month

# Compute the mean for each month of the year
mean_by_month = temp.groupby(index_month).mean()

# Plot the mean for each month of the year
ax = mean_by_month.plot(color='green',linewidth=2)

# Add a  horizontal line
ax.axhline(mean_by_month.max().max(), color='black', linestyle='--')
ax.axhline(mean_by_month.mean().max(), color='yellow', linestyle='--')
ax.axhline(mean_by_month.min().min(), color='red', linestyle='--')

# Specify the fontsize on the legend
plt.legend(fontsize=10)
plt.title("Positives by month data",color='red')
plt.xlabel("Months of a year",color='red')
# Show plot
plt.show()
```

<p>The Lowest Positives are received in the month of August and the maximum positives are received during the month of November</p>

```python
temp = df[["Total_Sentiment"]]
temp['Year'] = temp.index.year

mean_by_year = temp.groupby('Year')['Total_Sentiment'].mean()
```

```python
mean_by_year
```

```python
# Import statsmodels.api as sm
import statsmodels.api as sm

# Perform time series decompositon
decomposition = sm.tsa.seasonal_decompose(mean_by_year,model = 'additive',period=1)

# Print the seasonality component
print(decomposition.seasonal)
```

```python
# Extract the trend component
trend = decomposition.trend

# Plot the values of the trend
ax = trend.plot(figsize=(12, 6), fontsize=6)

# Specify axis labels
ax.set_xlabel('Date', fontsize=10)
ax.set_title('Seasonal component of Total Sentiment', fontsize=10)
plt.show()
```
